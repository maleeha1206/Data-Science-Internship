<h4> Problem Statement: </h4> Given the raw biological measurements of iris flowers, how can we use basic statistics and charts to understand their structural differences and find underlying data patterns.

<h1> Graphs Using seaborn </h1> 

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris= sns.load_dataset('iris') #Load the iris dataset
```

```python
print("Dataset Shape= ", iris.shape) #Display number of rows and columns in the dataset 
print("Dataset Columns= ", iris.columns) #Display dataset columns
iris.head() #Displays first 5 rows (If the number of rows is not mentioned then by default it displays 5 rows)
```

```python
sns.scatterplot(data=iris, x='petal_length', y='petal_width', hue='species') #Scatter Plot for relationship between variables. 
plt.title('Petal Length vs Petal Width')
```

```python
sns.histplot(data=iris, x='sepal_length', kde=True, hue='species', element='step') #Histogram for data Distribution
plt.title('Distribution of Sepal Length')
```

```python
sns.boxplot(data=iris, x='species', y='sepal_width') #Boxplot for outliers and spread of values
plt.title('Sepal Width Spread by Species')
```

<h1> Graphs Using matplotlib </h1>

```python
plt.scatter(iris['petal_length'], iris['petal_width'], color='purple', alpha=0.5)
plt.title('Scatter Plot: Petal Length vs Petal Width')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.grid(True, linestyle='--', alpha=0.5)  # Adds a nice background grid
```

```python
plt.hist(iris['sepal_length'], color='pink', edgecolor='purple')
plt.title('Histogram: Distribution of Sepal Length')
plt.xlabel('Sepal Length')
plt.ylabel('Frequency (Count)')
```

```python
# Matplotlib needs us to separate the data by species manually for a box plot
# Separating by species to look at the spread side-by-side
setosa_data = iris[iris['species'] == 'setosa']['sepal_width']
versicolor_data = iris[iris['species'] == 'versicolor']['sepal_width']
virginica_data = iris[iris['species'] == 'virginica']['sepal_width']

# Pass the separated data as a list to plt.boxplot
plt.boxplot([setosa_data, versicolor_data, virginica_data], tick_labels=['Setosa', 'Versicolor', 'Virginica']) #tick_labels assigns a name to the bottom of each box. The first label goes to the first list, the second label to the second list, and so on.

plt.title('Box Plot: Sepal Width Spread by Species')
plt.xlabel('Species')
plt.ylabel('Sepal Width')
```

<h3> Dataset Understanding and Description </h3>
Using Python's structural commands, we uncovered the layout of the data:

.shape Results: The dataset contains exactly 150 rows (individual flowers) and 5 columns (tracked features).

.columns Results: The columns consist of four physical measurements (sepal_length, sepal_width, petal_length, petal_width) and one category label (species).

.head() Results: Reveled that the data is organized neatly in a table where each row tracks a specific flower sample alongside its exact measurements.

<h3> Exploratory Data Analysis (EDA) with Graphs </h3> 
<h5> Scatter Plot Insight (Relationships): </h5> By plotting Petal Length against Petal Width, we notice a strong, upward diagonal trend. This proves that as an iris flower's petals grow longer, they consistently grow wider as well.

<h5> Histogram Insight (Distribution): </h5> The histogram for Sepal Length shows a wide, spread-out layout with a major peak hovering around 5.5 to 6.5 cm. This reveals that "average-sized" sepals are highly common, while exceptionally short or long sepals are quite rare.

<h5> Box Plot Insight (Spread & Outliers): </h5> The box plot highlights that the Setosa species generally has wider sepals than the other two species. Looking closely at the Virginica box plot, a lone dot rests well outside the main whisker lines—successfully flagging an outlier (a flower with an unusually small sepal width compared to its species group).
